```python
### Setting the data path
data_path = r'C:\Data\LearnMLAI\Applied-Geospatial-Data-Science-with-Python\Data'
```

```python
import pandas as pd
import geopandas as gpd
from geopandas.tools import geocode

# 1. Read in the raw CSV data
dc_attractions = pd.read_csv(data_path + r'\DC_Attractions.csv')

# 2. Geocode addresses using the free, open-source 'nominatim' provider
# We swap 'api_key' for a descriptive 'user_agent' string and keep a safe timeout
dc_attractions_gpd = geocode(
    dc_attractions['Address'], 
    provider='nominatim', 
    user_agent='dc_attractions_urban_analysis', 
    timeout=5
)

# 3. Join the attraction names back to the spatial coordinates
dc_attractions_gpd = dc_attractions_gpd.join(dc_attractions[['Attraction']])

# 4. Extract geometry coordinates into explicit latitude and longitude columns
dc_attractions_gpd['lon'] = dc_attractions_gpd['geometry'].x
dc_attractions_gpd['lat'] = dc_attractions_gpd['geometry'].y

# 5. Preview the finished spatial dataframe
dc_attractions_gpd.head()
```

```python
from ipyleaflet import Map, basemaps

DC_Map = Map(
    basemap=basemaps.CartoDB.Positron,
    center=(38.89951498583087, -77.03599825749647),
    zoom=12
)

DC_Map
```

```python
from ipyleaflet import Map, basemaps

DC_Map = Map(
    basemap=basemaps.OpenStreetMap.Mapnik,
    center=(38.89951498583087, -77.03599825749647),
    zoom=12
)

DC_Map
```

```python
# Mapping the attractions

for (index, row) in dc_attractions_gpd.iterrows():
    marker = Marker(location = [row.loc['lat'], row.loc['lon']], title=row.loc['Attraction'])
    DC_Map.add_layer(marker)
```
